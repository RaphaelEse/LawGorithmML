```python
import pandas as pd
import torch
from transformers import pipeline

csv_path = "/content/drive/MyDrive/Colab Notebooks/billsWithID_cleaned.csv"
df = pd.read_csv(csv_path)

# STEP 5: Load zero-shot classification model
device = 0 if torch.cuda.is_available() else -1
classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli", device=device)

# STEP 6: Define candidate labels
candidate_labels = ["Precedent", "Authority", "Definition", "Example", "Exception", "Amendment"]

# STEP 7: Run classification
df["high_confidence"] = df["context"].apply(
    lambda x: classifier(x, candidate_labels)["labels"][0] if isinstance(x, str) and x.strip() else ""
)

# STEP 8: Save to CSV
output_path = "/content/drive/MyDrive/Colab Notebooks/billsWithHighConfidence.csv"
df.to_csv(output_path, index=False)
print(f"Saved to {output_path}")
```
